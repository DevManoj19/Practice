```python
import scipy.stats as stats

# Data
A_scores = [80, 78]  # Exam scores for method A
B_scores = [85, 82]  # Exam scores for method B
C_scores = [90]      # Exam scores for method C

# Perform ANOVA
f_statistic, p_value = stats.f_oneway(A_scores, B_scores, C_scores)

# Print results
print("F-statistic:", f_statistic)
print("P-value:", p_value)

```

```python
import scipy.stats as stats
import pandas as pd

# Data
data = {
    'Temperature': [20, 25, 30, 35, 40, 45, 50],
    'Humidity': [40, 60, 45, 75, 55, 85, 65],
    'Crop_Yield': [100, 120, 110, 130, 115, 140, 125]
}

df = pd.DataFrame(data)

# Perform two-way ANOVA
formula = 'Crop_Yield ~ C(Temperature) * C(Humidity)'
model = stats.(formula, df).fit()
anova_table = stats.anova_lm(model, typ=2)

# Print the ANOVA table
print(anova_table)
```
